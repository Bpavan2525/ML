18. What is an optimizer? What are the different types of optimizers? Explain each with an example.
- an optimizer adjusts the parameters like weights, bias to increase the model's performance.
- Gradient Descent: it helps the algorithm to slowly updates the model's parameter in the direction of steepest or ascent of the loss function ex: hiker trying to reach the bottom of the valley.
- Stochastic Gradient Descent: it iterates over a randomly selected data point and updates the parameter based on the gradient calculated but makes more noise. ex: instead of selecting all the data points it selects the random data point to compute the gradient.
- Adam: it is an adaptive learning optimizer which uses the advantages of rmsprop and momentum ex: it uses moving averages of past gradients and their squares to adjust the learning of each parameter.
- RMSprop: it is also similar to adam but it uses moving average of gradients ex: it maintains the exponential square of past gradients and normalize the current gradient.
- Adagrad: it adapts the learning rate to optimize each parameter individually and takes the squares of past gradients and adjusts the learning rate. ex: parameters which are adjusted recently have a larger learning rate and which are from past will have a smaller elarning rate.
